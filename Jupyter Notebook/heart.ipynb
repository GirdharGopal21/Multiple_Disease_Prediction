```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Downloads/heart.csv")
df.head()
```

```python
df['target'].value_counts()
```

```python
cor = df.corr()
cor
```

```python
df.groupby('target').mean()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
print(x.shape)
print(y.shape)
```

```python
# stratify=y is used for even distributn of data so that model gets trained evenly for both 1 and 0
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y,random_state=2,test_size=0.2,stratify=y)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.svm import SVC
clf1 = SVC()
clf1.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
Y_test_pred = clf1.predict(X_test)
```

```python
print(accuracy_score(Y_test,Y_test_pred))
```

```python
# using naive bayes classifier
from sklearn.naive_bayes import GaussianNB
clf3 = GaussianNB()
clf3.fit(X_train,Y_train)
Y_test_pred3 = clf3.predict(X_test)
print(accuracy_score(Y_test,Y_test_pred3))
```
