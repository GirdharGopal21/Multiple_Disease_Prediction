```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("C:/Users/Divyansh/Downloads/brain_stroke.csv")
df.head()
```

```python
df.info()
```

```python
df['gender'].unique()
le1 = LabelEncoder()
df['gender'] = le1.fit_transform(df['gender'])
df['gender'].unique()
df.head()
#male-->1
#female-->0
```

```python
#for ever_married
le = LabelEncoder()

# Get unique categories before encoding
original_categories = df['ever_married'].unique()

# Fit and transform the data
df['ever_married'] = le.fit_transform(df['ever_married'])

# Create a mapping of numeric values to original categories
numeric_to_category_mapping = dict(zip(le.transform(original_categories), original_categories))

# Print the mapping
print("Numeric value to category mapping:")
for numeric_value, category in numeric_to_category_mapping.items():
    print(f"{numeric_value}: {category}")
df.head()
```

```python
# work type
# Get unique categories before encoding
original_categories = df['work_type'].unique()

# Fit and transform the data
df['work_type'] = le.fit_transform(df['work_type'])

# Create a mapping of numeric values to original categories
numeric_to_category_mapping = dict(zip(le.transform(original_categories), original_categories))

# Print the mapping
print("Numeric value to category mapping:")
for numeric_value, category in numeric_to_category_mapping.items():
    print(f"{numeric_value}: {category}")
df.head()
```

```python
#Residence_type
# Get unique categories before encoding
original_categories = df['Residence_type'].unique()

# Fit and transform the data
df['Residence_type'] = le.fit_transform(df['Residence_type'])

# Create a mapping of numeric values to original categories
numeric_to_category_mapping = dict(zip(le.transform(original_categories), original_categories))

# Print the mapping
print("Numeric value to category mapping:")
for numeric_value, category in numeric_to_category_mapping.items():
    print(f"{numeric_value}: {category}")
df.head()
```

```python
df.info()
print(df['stroke'].unique())
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:,-1]
print(x.shape)
print(x.info())
print(y.shape)
print(y.info())
print(y.unique())
```

```python
# Splitting the data
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2,random_state=4,stratify=y)
print(X_train.shape)
print(Y_train.shape)
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(X_train,Y_train)
```

```python
y_test_pred = clf.predict(X_test)
score = accuracy_score(Y_test,y_test_pred)
print(score)
```

```python
#Naive bayes classifier
from sklearn.naive_bayes import GaussianNB
clf2 = GaussianNB()
clf2.fit(X_train,Y_train)
y_test_pred2 = clf2.predict(X_test)
score = accuracy_score(Y_test,y_test_pred2)
print(score)
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
clf3 = RandomForestClassifier()
clf3.fit(X_train,Y_train)
y_test_pred3 = clf3.predict(X_test)
score = accuracy_score(Y_test,y_test_pred3)
print(score)
```

```python
#KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
clf4 = KNeighborsClassifier()
clf4.fit(X_train,Y_train)
y_test_pred4 = clf4.predict(X_test)
score = accuracy_score(Y_test,y_test_pred4)
print(score)
```
